# Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plotting style
sns.set(style="whitegrid")

# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Function to plot boxplots for numerical columns
def plot_boxplots(df, cols):
    plt.figure(figsize=(15, 15))
    for i, col in enumerate(cols, 1):
        plt.subplot(len(cols)//3 + 1, 3, i)
        sns.boxplot(y=df[col], color='lightblue', fliersize=5)
        plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()

# Call the function with your dataset
plot_boxplots(df, numerical_cols)
```

```python

# Set the plotting style
sns.set(style="whitegrid")

# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Plot probability distribution (KDE) and histogram
def plot_numerical_distributions(df, cols):
    plt.figure(figsize=(15, 15))
    for i, col in enumerate(cols, 1):
        plt.subplot(len(cols)//3 + 1, 3, i)
        sns.histplot(df[col].dropna(), kde=True, bins=30, color='blue')  # kde=True for probability distribution
        plt.title(f'Probability & Frequency Distribution of {col}')
    plt.tight_layout()
    plt.show()

# Call the function with your dataset
plot_numerical_distributions(df, numerical_cols)
```

```python

# Select the Gender column for countplot
categorical_col = 'Gender'

# Function to generate countplot with observations for Gender
def plot_gender_distribution(df, col):
    plt.figure(figsize=(8, 5))
    
    # Generate countplot
    ax = sns.countplot(x=df[col], order=df[col].value_counts().index, palette='Set2')
    plt.xticks(rotation=45)
    plt.title(f'Countplot of {col}')

    # Annotate bars with counts and percentages
    total = len(df[col])
    for p in ax.patches:
        percentage = f'{100 * p.get_height() / total:.2f}%'
        ax.annotate(f'{p.get_height()} ({percentage})', (p.get_x() + p.get_width() / 2, p.get_height()), 
                    ha='center', va='bottom', fontsize=10, color='black')

    # Print value counts
    counts = df[col].value_counts()
    print(f"\n--- {col} Frequency Distribution ---")
    print(counts)

    # Observation
    if 'Male' in counts and 'Female' in counts:
        more_male = counts['Male'] > counts['Female']
        print("Observation:", "More male candidates." if more_male else "More female candidates.")
    else:
        print("Observation: Gender categories may vary.")

    plt.tight_layout()
    plt.show()

# Call the function to generate the plot
plot_gender_distribution(df, categorical_col)
```

```python

# Select the column for analysis
job_city_col = 'JobCity'

# Function to generate countplot for top 10 Job Cities
def plot_top_cities(df, job_city_col):
    plt.figure(figsize=(10, 6))

    # Get the top 10 job cities
    top_cities = df[job_city_col].value_counts().nlargest(10).index

    # Generate countplot for top 10 Job Cities
    sns.countplot(x=job_city_col, data=df[df[job_city_col].isin(top_cities)], 
                  order=top_cities, palette='Set2')
    plt.xticks(rotation=45)
    plt.title('Top 10 Job Cities')

    # Annotate Job City bars with counts and percentages
    total_cities = len(df[df[job_city_col].isin(top_cities)])
    for p in plt.gca().patches:
        percentage = f'{100 * p.get_height() / total_cities:.2f}%'
        plt.annotate(f'{p.get_height()} ({percentage})', 
                     (p.get_x() + p.get_width() / 2, p.get_height()), 
                     ha='center', va='bottom', fontsize=10, color='black')

    plt.tight_layout()
    plt.show()

# Call the function to generate the plot
plot_top_cities(df, job_city_col)
```
